```python
df1 = data
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

def get_score (row):
   return sia.polarity_scores(row['text'])['neg']

df1['score'] = df1.apply (lambda row: get_score(row), axis=1)
```
